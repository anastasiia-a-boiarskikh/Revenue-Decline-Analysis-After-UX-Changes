### Дерево метрик


Целевой показатель: выручка (Revenue). Основную выручку в продукте приносят продажи предметов через внутриигровой магазин

Первый уровень метрик, Revenue = AU × ARPAU:
- Active Users (AU, кол-во активных пользователей). Кол-во активных пользователей — это количество пользователей, которые запускают приложение с игрой в определённый период
- Average Revenue per Active User (ARPAU, средний доход, который приносит продукту активный пользователь). ARPAU — это средний доход от внутриигрового магазина на одного пользователя, который запускал приложение с игрой в определённый промежуток времени. Рассчитывается как ARPAU = выручка за период / количество пользователей с запусками приложения за период

Второй уровень метрик:
- Активную аудиторию можно представить как совокупность новых пользователей и удержанных пользователей
- Средний доход на активного пользователя зависит от конверсии в покупку и среднего чека во внутриигровом магазине

Третий уровень метрик:
- Новые пользователи приходят из органического или платного канала
- На кол-во удержанных пользователей вляет метрика retention rate (RR), которая зависит от качества пользовательского опыта
- Конверсия в покупку сильно зависит от метрики «Доля пользователей, у которых пройдено более 10 уровней»
- Средний чек во внутрииигровом магазине зависит от средней стоимости предмета и среднего количества предметов в заказе

Четвертый уровень метрик:
- Одна из главных метрик качества пользовательского опыта — «Средняя доля пройденных уровней на сессию». Она показывает, сколько уровней от общего количества запущенных успешно пройдено пользователем за одну игровую сессию. 

### Гипотезы о причинах снижения выручки
1. Новый алгоритм генерации уровней предлагает слишком сложные уровни в начале. Пользователи не успевают пройти достаточно уровней, чтобы "залипнуть" и начать совершать покупки во внутриигровом магазине. По дереву: снижается метрика "Доля пользователей, у которых пройдено более 10 уровней" -> снижается конверсия в покупку -> снижается средняя выручка на активного пользователя -> снижается выручка
2. Из-за редизайна пользователи не могут найти нужные им вещи в игровом магазине, поэтому тратят больше времени на поиск и покупают меньше. По дереву: снижается кол-во предметов в заказе -> снижается средний чек заказа по внутриигровом магазине -> снижается средняя выручка на активного пользователя -> снижается выручка
3. Новый алгоритм генерирует значительно более сложные уровни, чем отталкивает удержанных пользователей. По дереву: снижается средняя доля пройденных уровней на сессию -> снижается RR -> снижается кол-во удержанных пользователей -> снижается кол-во активных пользователей -> снижается выручка

Дополнительные гипотезы из эталонного решения:<br>
4. Ухудшилась стратегия привлечения новых пользователей. По дереву: снизилось кол-во новых пользователей из платных или реферальных каналов -> снизилось кол-во новых пользователей -> снизилось кол-во активных пользователей -> снизилась выручка<br>
5. Пользователи стали покупать более дешевые предметы во внутриигровом магазине. По дереву: снижается средняя стоимость предметов -> снижается средний чек заказа во внутриигровом магазине -> снижается средняя выручка на активного пользователя -> снижается выручка

### Метрики для проверки каждой из гипотез
1. Пользователи стали менее активно совершать покупки (снизилась конверсия в покупку):
    - воронка конверсии от запуска приложения до покупки предметов во внутриигровом магазине
    - доля пользователей, у которых пройдено более 10 уровней<br>
<br>
2. Пользователи стали покупать меньше товаров:
    - среднее кол-во предметов в заказе<br>
<br>
3. Ухудшение опыта текущих пользователей: 
    - retention rate
    - cредняя доля пройденных уровней на сессию<br>
<br>
4. Ухудшение стратегии привлечения новых пользователей:
    - кол-во новых пользователей из оганического / платного / реферального каналов<br>
<br>
5. Пользователи стали покупать более дешевые товары:
    - средняя стоимость предметов в заказе<br>
    
Для гипотез 3, 4 также полезно рассчитать MAU

#### 1. Получение датасета для расчета 
- динамики MAU по платформам и странам пользователей
- динамики кол-ва новых пользователей по отдельным каналам привлечения
- динамики retention rate по когортам (разделение на когорты по месяцу первого взаимодействия с приложением)

Датасет должен содержать:
- уникальный идентификатор пользователя
- месяц первого взаимодейтсвия с приложением
- все месяцы взаимодействия с приложением
- платформу, страну и канал привлечения пользователя

Т.к. учитываем  как зарегистрированных, так и не зарегистрированных пользователей, вместо `user_id` воспользуемся полем `device_id`

```
WITH first_month AS (
    SELECT
        device_id,
        MIN(DATE_TRUNC('month', time)::date) AS first_activity_month
    FROM mines.events
    WHERE event_name = 'app_start'
    GROUP BY device_id
),

all_month AS (
    SELECT
        device_id,
        DATE_TRUNC('month', time)::date AS all_activity_month
    FROM mines.events
    WHERE event_name = 'app_start'
    GROUP BY 1, 2
)

SELECT
    f.device_id,
    f.first_activity_month,
    a.all_activity_month,
    u.platform,
    u.country_code,
    u.traffic_type
FROM first_month AS f
LEFT JOIN all_month AS a USING(device_id)
LEFT JOIN mines.users AS u USING(device_id);
```

````python
df['n_period'] = (
    df['all_activity_month'].dt.to_period('M').astype('int64')
    - df['first_activity_month'].dt.to_period('M').astype('int64')
)

users_all_cohorts = (
    pd.pivot_table(
        data=df,
        index='first_activity_month',
        values='device_id',
        columns='n_period',
        aggfunc=pd.Series.nunique,
    )
)

# каждое значение строки датафрейма делим 
# на значение в периоде 0 с помощью функции div()
rr_all_cohorts = 100 * users_all_cohorts.div(users_all_cohorts[0], axis=0)

# для корректного отображения индекса приведём его к типу ('%Y-%m-%d')
rr_all_cohorts.index = pd.to_datetime(rr_all_cohorts.index).strftime('%Y-%m-%d')

cmap = mcolors.LinearSegmentedColormap.from_list('green_red', ['#FF3333', '#FFFF66', '#4C9900'], N=100)

plt.subplots(figsize=(15,9))

ax = sns.heatmap(
    data=rr_all_cohorts,
    cmap=cmap,
    annot=True,
    fmt=".0f",
    linewidth=0.5,
    annot_kws={"size": 16},
    cbar=True,
)

cbar = ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=13)

plt.title('Классический Retention по когортам в %', fontsize=16)
plt.xlabel('Период', size=14, labelpad=10)
plt.ylabel('Когорта', size=14, labelpad=20)
plt.tick_params(labelsize=13) # чтобы сделать подписи данных на оси Y (когорты пользователей) немного крупнее
plt.savefig('retention_rate_cohort_analysis.png')
````

### Исследование ветви "Средняя выручка на активного пользователя"

````python
transactions = pd.read_csv('transactions.csv')
events = pd.read_csv('events.csv')

transactions['dt'] = pd.to_datetime(transactions['time']).dt.to_period('M').dt.to_timestamp()
events['dt'] = pd.to_datetime(events['time']).dt.to_period('M').dt.to_timestamp()

display(transactions.head())
display(events.head())
````

````python
# расчет ARPAU
success_transactions = transactions[transactions.status=='success']
transactions_sum = success_transactions.groupby('dt')['price'].sum().reset_index()

users_n = events[events.event_name=='app_start'].groupby('dt')['device_id'].nunique().reset_index()

merged_df = pd.merge(
    transactions_sum,
    users_n,
    how='left',
    on='dt',
)

merged_df['ARPAU'] = merged_df.price / merged_df.device_id
display(merged_df)

plt.figure(figsize=(10,6))

plt.plot(
    merged_df.dt, 
    merged_df.ARPAU,
    marker='o'
)

plt.title('Средняя выручка на пользователя по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Средняя выручка на пользователя')
plt.grid()
plt.savefig('ARPAU_monthly.png')
````

````python
# расчет воронки конверсии в покупку 
# по этапам store_open, item_add_to_cart
# checkout_load, order_completed

# работаем с датасетами events, users

events['time'] = pd.to_datetime(events['time'])
users['install_time'] = pd.to_datetime(users['install_time'])

df = pd.merge(
    events, 
    users[['device_id', 'install_time']], 
    how='left', 
    on='device_id'
)

# фильтруем события в пределах окна атрибуции
attribution_window = 30

df['diff_days'] = (df['time'] - df['install_time']).dt.days
df = df[df['diff_days'].between(0, attribution_window)]

df['install_month'] = df.install_time.dt.to_period('M').dt.to_timestamp()

event_list = ['store_open', 'item_add_to_cart', 'checkout_load', 'order_completed']
funnel_dynamic = df.groupby('install_month')['device_id'].nunique().reset_index(name='install_devices')
    
plt.figure(figsize=(10, 6))

for event in event_list:
    df_event = df[df['event_name'] == event]
    df_event = df_event.groupby('install_month')['device_id'].nunique().reset_index(name=event)
    funnel_dynamic = funnel_dynamic.merge(df_event, on='install_month', how='left')
    funnel_dynamic[event] = (funnel_dynamic[event] / funnel_dynamic['install_devices']) * 100

    plt.plot(funnel_dynamic['install_month'], funnel_dynamic[event], marker='o', label=event)

plt.title('Динамика конверсии воронки (окно атрибуции 30 дней)')
plt.xlabel('Месяц установки приложения')
plt.ylabel('Конверсия (%)')
plt.xticks(rotation=45)
plt.legend(event_list)
plt.grid(True)
plt.tight_layout()
plt.savefig('funnel_dynamic_monthly.png')

display(funnel_dynamic)
````

````python
# расчет метрики средняя цена
# предметов в заказе

# работаем с датасетом transactions

success_transactions = transactions[transactions['status'] == 'success']
success_transactions['month'] = pd.to_datetime(success_transactions['time']).dt.to_period('M').dt.to_timestamp()

grouped_transactions = (
    success_transactions
        .groupby('month')
        .agg(
            price_sum=('price', 'sum'),
            item_quantity=('transaction_id', 'count')
        )
        .reset_index()
)

grouped_transactions['avg_item_price'] = (
    grouped_transactions['price_sum'] 
    / grouped_transactions['item_quantity']
)

plt.figure(figsize=(10, 6))

plt.plot(
    grouped_transactions['month'], 
    grouped_transactions['avg_item_price'], 
    marker='o'
)

plt.title('Средняя цена предметов в заказе по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Средняя цена предметов в заказе')
plt.grid()
plt.ylim([20, 26])
plt.savefig('avg_item_cost_monthly.png')
````

````python
# расчет метрики среднее кол-во
# предметов в заказе

# работаем с датасетом transactions

success_transactions = transactions[transactions['status'] == 'success']
success_transactions['month'] = pd.to_datetime(success_transactions['time']).dt.to_period('M').dt.to_timestamp()
grouped_transactions = (
    success_transactions
        .groupby('month')
        .agg(
           order_quantity=('transaction_id', 'nunique'),
           item_quantity=('transaction_id', 'count')
        )
        .reset_index()
)

grouped_transactions['avg_item_count'] = (
    grouped_transactions['item_quantity']
    / grouped_transactions['order_quantity']
)

display(grouped_transactions)

plt.figure(figsize=(10,6))

plt.plot(
    grouped_transactions.month,
    grouped_transactions.avg_item_count,
    marker='o'
)

plt.title('Среднее кол-во предметов в заказе по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Среднее кол-во предметов в заказе')
plt.ylim([1, 2])
plt.grid()
plt.savefig('avg_item_quantity_monthly.png')
````

````python
# расчет доли пользователей, у которых
# пройдено более 10 уровней

# используем датасеты game sessions и users

game_sessions['month'] = pd.to_datetime(game_sessions['time_end']).dt.to_period('M').dt.to_timestamp()
users['month'] = pd.to_datetime(users['install_time']).dt.to_period('M').dt.to_timestamp()

df_levels = game_sessions[
    (game_sessions.level_id == 11)
    & (game_sessions.success_fact == 1)
]

df_levels_grouped = (
    df_levels
        .groupby('month')
        .agg({'device_id': 'nunique'})
        .reset_index()
)

df_levels_grouped['cumsum_users_level_success'] = (
    df_levels_grouped['device_id'].cumsum()
)

df_users_grouped = (
    users
        .groupby('month')
        .agg({'device_id': 'nunique'})
        .reset_index()
)
df_users_grouped['cumsum_users_all'] = (
    df_users_grouped['device_id'].cumsum()
)

table_final = pd.merge(
    df_levels_grouped,
    df_users_grouped,
    how='inner',
    on='month',
)

table_final['level_10_finish_rate'] = round(100 * (
    table_final.cumsum_users_level_success
    / table_final.cumsum_users_all
), 2)

display(table_final[[
    'month', 
    'cumsum_users_level_success',
    'cumsum_users_all',
    'level_10_finish_rate'
]])

plt.figure(figsize=(10,6))

plt.plot(
    table_final['month'],
    table_final['level_10_finish_rate'],
    marker='o'
)

plt.title('Доля пользователей, у которых пройдено более 10 уровней')
plt.xlabel('Месяц')
plt.ylabel('Пользователей, %')
plt.grid()

plt.savefig('10_level_passed_share_monthly.png')
````
